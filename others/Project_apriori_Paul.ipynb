```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import numpy as np
```

```python

movies = pd.read_csv("movies.txt", on_bad_lines='skip', sep = ';', header = None)
```

```python
movies = movies.iloc[1: , :]
```

```python
movies
```

```python

mv = []
for i in range(0, movies.shape[0]):
  for j in range(0, movies.shape[1]):
    mv.append(movies.values[i, j])
mv = np.array(mv)
```

```python
mv_df = pd.DataFrame(mv, columns = ['title'])

mv_df['rate count'] = 1
```

```python
indexNames = mv_df[mv_df['title'] == 'nan'].index
mv_df.drop(indexNames, inplace = True)
```

```python
df_table = mv_df.groupby('title').sum().sort_values('rate count', ascending = False).reset_index()
```

```python
conv = []
for i in range(movies.shape[0]):
  conv.append([str(movies.values[i, j]) for j in range(movies.shape[1])])

conv = np.array(conv)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt


conv_te = TransactionEncoder()
```

```python
conv_te_ary = conv_te.fit(conv).transform(conv)
dataset = pd.DataFrame(conv_te_ary, columns = conv_te.columns_)
```

```python
def encode_units(x):
  if x == False:
    return 0
  if x == True:
    return 1
  
dataset = dataset.applymap(encode_units)
```

```python
dataset = dataset.iloc[: , :-1]
dataset
```

```python
frequent_movies = apriori(dataset, min_support = 0.05, use_colnames = True)
```

```python
frequent_movies['length'] = frequent_movies['itemsets'].apply(lambda x: len(x))
```

```python
rules = association_rules(frequent_movies, metric = 'lift', min_threshold = 1.2)
rules
```

```python
rules['antecedents_length'] = rules['antecedents'].apply(lambda x: len(x))
rules['consequents_length'] = rules['consequents'].apply(lambda x: len(x))
rules.sort_values('lift', ascending = False)
```
